```python
import matplotlib.pyplot as plot

# allow plots to show in Jupyter Notebook
%matplotlib inline
 
import numpy as np
import pandas as pd

# Autoreload notebook when changes are made
%reload_ext autoreload
%autoreload
```

```python
# Create x and y arrays
# x1 = Create an array of 10 floats from 0 to 5
x1 = np.linspace(0, 5, 10)
y1 = x1**2

plot.plot(x1,y1) # Display plot
plot.title("x² Graph")  # Plot title

plot.xlabel("x")
plot.ylabel("x²")
```

```python
plot.subplot(1, 2, 1)  # (row, column, plot num)
plot.plot(x1, y1, "r") # "r" - make plot line red

plot.subplot(1, 2, 2)
plot.plot(x1, y1, "b")
```

```python
# Figure objects contain all plot elements
# Can hold many axes objects, but a given axes object can only exist in one figure
# Figure is essentially the paper, axes are the graphs or plotlines

# figsize = define width and height in inches
# dpi = res of figure in dots per inch
fig1 = plot.figure(figsize=(5,4), dpi=100)

# add an axes object with [left, bottom, width, height] in range 0-1
axis1 = fig1.add_axes([0.1, 0.1, 0.9, 0.9])

axis1.set_xlabel("x")
axis1.set_ylabel("y")
axis1.set_title("x² graph")

# Display plot on the axes
axis1.plot(x1, y1, label = "x/x²")

# Second plotline on the same axes
axis1.plot(y1, x1, label = "x²/x")

# Add optional legend with location number
# best guess: 0, upper right: 1, upper left: 2, lower left: 3, lower right: 4
# 0 makes it look for locations with minimal overlap. Might lead to slowdown on complex graphs
axis1.legend(loc=0)

# Create a second axis inside the figure
axis2 = fig1.add_axes([0.45,0.45,0.4,0.3])
axis2.set_xlabel("x")
axis2.set_ylabel("y")
axis2.set_title("Small x² graph")
axis2.plot(x1,y1,"r")

# Add text to plot (x, y, "text")
# (0,0) is the centre of the figure
axis2.text(0, 40, 'Message')
```

Subplots

```python
# How to define multiple plots with subplots
# axis2 is a list of generated axes objects
# create a set of subplots based on nrows,ncols
# can assign shared values between plots
fig2, axis2 = plot.subplots(figsize= (8,4), nrows=1, ncols=3)
plot.tight_layout() # generate space between plots

# access the different axes objects with list indices
axis2[0].set_title("Plot 1")
axis2[1].set_title("Plot 2")
axis2[2].set_title("Plot 3")

axis2[1].set_xlabel("x")
axis2[1].set_ylabel("x²")
axis2[1].plot(x1,y1)
```

Controlling the Appearance

```python
fig3 = plot.figure(figsize=(6,4))
axis3 = fig3.add_axes([0,0,1,1])

# Default colours = blue(b), green(g), red(r), cyan(c), magenta(m), yellow(y), black(k), white(w)
# color="0.75" - mix with 75% grey
# Can use hexcode or dulux colour names like limegreen or royalblue
# https://en.wikipedia.org/wiki/Web_colors
# aplha variable defines opacity
# lw = line width. default 1
# Line styles - matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html

# Markers - exactly what it sounds like - https://matplotlib.org/3.3.0/api/markers_api.html
# markerfacecolor - fill colour
# markeredgecolor - stroke colour
# markeredgewidth - stroke size

axis3.plot(x1,y1,color='navy', alpha=.75, lw=2, ls='-.',
           marker='o', markersize=7, markerfacecolor='y',
           markeredgecolor='y', markeredgewidth=4)

# Set bounds
axis3.set_xlim([0,3])
axis3.set_ylim([0,8])

# add grid, colour, dashes
axis3.grid(True, color="0.6", dashes=(5,2,1,2))
axis3.set_facecolor("#cdf2cc")
```

Save Graph to a file

```python
fig3.savefig("fig3.png")
```
