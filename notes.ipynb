```python
import matplotlib.pyplot as plot

# allow plots to show in Jupyter Notebook
%matplotlib inline
 
import numpy as np
import pandas as pd

# Autoreload notebook when changes are made
%reload_ext autoreload
%autoreload
```

```python
# Create x and y arrays
# x1 = Create an array of 10 floats from 0 to 5
x1 = np.linspace(0, 5, 10)
y1 = x1**2

plot.plot(x1,y1) # Display plot
plot.title("xÂ² Graph")  # Plot title

plot.xlabel("x")
plot.ylabel("xÂ²")
```

```python
plot.subplot(1, 2, 1)  # (row, column, plot num)
plot.plot(x1, y1, "r") # "r" - make plot line red

plot.subplot(1, 2, 2)
plot.plot(x1, y1, "b")
```

```python
# Figure objects contain all plot elements
# Can hold many axes objects, but a given axes object can only exist in one figure
# Figure is essentially the paper, axes are the graphs or plotlines

# figsize = define width and height in inches
# dpi = res of figure in dots per inch
fig1 = plot.figure(figsize=(5,4), dpi=100)

# add an axes object with [left, bottom, width, height] in range 0-1
axis1 = fig1.add_axes([0.1, 0.1, 0.9, 0.9])

axis1.set_xlabel("x")
axis1.set_ylabel("y")
axis1.set_title("xÂ² graph")

# Display plot on the axes
axis1.plot(x1, y1, label = "x/xÂ²")

# Second plotline on the same axes
axis1.plot(y1, x1, label = "xÂ²/x")

# Add optional legend with location number
# best guess: 0, upper right: 1, upper left: 2, lower left: 3, lower right: 4
# 0 makes it look for locations with minimal overlap. Might lead to slowdown on complex graphs
axis1.legend(loc=0)

# Create a second axis inside the figure
axis2 = fig1.add_axes([0.45,0.45,0.4,0.3])
axis2.set_xlabel("x")
axis2.set_ylabel("y")
axis2.set_title("Small xÂ² graph")
axis2.plot(x1,y1,"r")

# Add text to plot (x, y, "text")
# (0,0) is the centre of the figure
axis2.text(0, 40, 'Message')
```

Subplots

```python
# How to define multiple plots with subplots
# axis2 is a list of generated axes objects
# create a set of subplots based on nrows,ncols
# can assign shared values between plots
fig2, axis2 = plot.subplots(figsize= (8,4), nrows=1, ncols=3)
plot.tight_layout() # generate space between plots

# access the different axes objects with list indices
axis2[0].set_title("Plot 1")
axis2[1].set_title("Plot 2")
axis2[2].set_title("Plot 3")

axis2[1].set_xlabel("x")
axis2[1].set_ylabel("xÂ²")
axis2[1].plot(x1,y1)
```

Controlling the Appearance

```python
fig3 = plot.figure(figsize=(6,4))
axis3 = fig3.add_axes([0,0,1,1])

# Default colours = blue(b), green(g), red(r), cyan(c), magenta(m), yellow(y), black(k), white(w)
# color="0.75" - mix with 75% grey
# Can use hexcode or dulux colour names like limegreen or royalblue
# https://en.wikipedia.org/wiki/Web_colors
# aplha variable defines opacity
# lw = line width. default 1
# Line styles - matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html

# Markers - exactly what it sounds like - https://matplotlib.org/3.3.0/api/markers_api.html
# markerfacecolor - fill colour
# markeredgecolor - stroke colour
# markeredgewidth - stroke size

axis3.plot(x1,y1,color='navy', alpha=.75, lw=2, ls='-.',
           marker='o', markersize=7, markerfacecolor='y',
           markeredgecolor='y', markeredgewidth=4)

# Set bounds
axis3.set_xlim([0,3])
axis3.set_ylim([0,8])

# add grid, colour, dashes
axis3.grid(True, color="0.6", dashes=(5,2,1,2))
axis3.set_facecolor("#cdf2cc")
```

Tables

```python
# format column data to 2 decimals
goog_data['Open'] = pd.Series([round(val, 2) for val in goog_data['Open']], 
                              index = goog_data.index)
goog_data['High'] = pd.Series([round(val, 2) for val in goog_data['High']], 
                              index = goog_data.index)
goog_data['Low'] = pd.Series([round(val, 2) for val in goog_data['Low']], 
                              index = goog_data.index)
goog_data['Close'] = pd.Series([round(val, 2) for val in goog_data['Close']], 
                              index = goog_data.index)
goog_data['Adj Close'] = pd.Series([round(val, 2) for val in goog_data['Adj Close']], 
                              index = goog_data.index)

# get last 5 days of stock data
stock_data = goog_data[-5:]
stock_data

# define headers
col_head = ('Date','Open','High','Low','Close','Adj Close','Volume')

stock_data_np = stock_data.to_numpy()
stock_data_np

# Add padding around cells in table
plot.figure(linewidth=2, tight_layout={'pad':.5}, figsize=(5,3))

# Get rid of axes and plot box
axes_8 = plot.gca()
axes_8.get_xaxis().set_visible(False)
axes_8.get_yaxis().set_visible(False)
plot.box(on=None)

# np.full returns an array filled with 0.1
# cm is a colormap object we are using to use a default blue color
# matplotlib.org/3.1.0/tutorials/colors/colormaps.html
ccolors = plot.cm.Blues(np.full(len(col_head), 0.2))

# Receives data, loc, list of column headers, column header color as array of colors
# You can also add rowLabel, rowColours, rowLoc: Text alignment
the_table = plot.table(cellText=stock_data_np, loc='center', colLabels=col_head,
                     colColours=ccolors)
# Set table font size
the_table.set_fontsize(14)
the_table.scale(3, 2.5)
```

3D Surfaces

```python
# Needed for creating 3D axes
from mpl_toolkits import mplot3d

fig_9 = plot.figure(figsize=(8,5),dpi=100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection='3d')

# Create a 3D scatterplot
# The darker points are represented that way to seem closer to you
z_3 = 15 * np.random.random(100)
x_3 = np.sin(z_3) * np.random.randn(100)
y_3 = np.cos(z_3) * np.random.randn(100)
# axes_9.scatter3D(x_3, y_3, z_3, c=z_3, cmap='Blues')

# You can create contour plots by defining a function for z based on x & y
def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x_4 = np.linspace(-6, 6, 30)
y_4 = np.linspace(-6, 6, 30)

# Creates a rectangular grid out of 2 given 1D arrays
x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4,y_4)

# Change viewing angle to reorient camera 60 degrees and rotate 55
axes_9.view_init(45,55)

# Provide x, y, z, contours and color map
# axes_9.contour3D(x_4,y_4,z_4,80,cmap='Blues')
axes_9.set_xlabel('x')
axes_9.set_ylabel('y')
axes_9.set_zlabel('z')

# You can create wireframes
# axes_9.plot_wireframe(x_4,y_4,z_4,color='blue')

# You can create surface plots which is wireframe with filled faces
axes_9.plot_surface(x_4,y_4,z_4, rstride=1, cstride=1,cmap='Blues',edgecolor='none')
```
